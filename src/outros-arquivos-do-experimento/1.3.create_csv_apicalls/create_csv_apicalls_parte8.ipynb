```python
import pandas as pd
import modin.pandas as mpd
import ray
import os


# ray.shutdown()
os.environ["MODIN_ENGINE"] = "ray"

runtime_env = {
    'env_vars': {
        "RAY_memory_monitor_refresh_ms": "100",
        "RAY_memory_usage_threshold": "0.85"
     }
}
ray.init(runtime_env=runtime_env)
```

```python
def is_valid_file(file_name, file_type):
    return file_name.endswith(file_type)

def is_permission_file(file_name):
    return is_valid_file(file_name, "-analysis-permissions.csv")

def is_apicall_file(file_name):
    return is_valid_file(file_name, "-analysis-apicalls.csv")

def is_opcode_file(file_name):
    return is_valid_file(file_name, "-analysis-opcodes.csv")
```

```python
files_completo = sorted(os.scandir("3.PREPROCESSING-40k/"), key=lambda e: e.name)
```

```python
files = files_completo[105_000:120_000]
```

```python
count = 0

total_files = len(files)
```

```python
output_csv = "csv/"
permission_df_list_tmp = []
permission_df_list = []
apicall_df_list_tmp = []
apicall_df_list = []
opcode_df_list_tmp = []
opcode_df_list = []

count = 0

df_opcodes = mpd.DataFrame()
df_permissions = mpd.DataFrame()
df_apicalls = mpd.DataFrame()

for file in files:
        count = count + 1
        
        print(chr(27) + "[2J")
        print(f"{count}/{total_files}")

        if is_permission_file(file.name):
             continue
            # df_tmp = mpd.read_csv(file.path)
            # df_tmp['file_name'] = file.name.split('-')[0]
            # permission_df_list_tmp.append(df_tmp)
        elif is_apicall_file(file.name):
            df_tmp = mpd.read_csv(file.path)
            # df_tmp['file_name'] = file.name.split('-')[0]
            apicall_df_list_tmp.append(df_tmp)
        elif is_opcode_file(file.name):
             continue
            # df_tmp = mpd.read_csv(file.path)
            # df_tmp['file_name'] = file.name.split('-')[0]
            # opcode_df_list_tmp.append(df_tmp)

        if len(apicall_df_list_tmp) == 1_000:
            print('Concatenando lote')
            apicall_df_list.append(mpd.concat(apicall_df_list_tmp, ignore_index=True))
            # permission_df_list.append(mpd.concat(permission_df_list_tmp, ignore_index=True))
            # opcode_df_list.append(mpd.concat(opcode_df_list_tmp, ignore_index=True))
            
            # count_group = 0
            apicall_df_list_tmp = []
            # permission_df_list_tmp = []
            # opcode_df_list_tmp = []



if len(apicall_df_list_tmp) > 0:
    print('Concatenando resto apicalls')
    apicall_df_list.append(mpd.concat(apicall_df_list_tmp, ignore_index=True))
    apicall_df_list_tmp = []

id = str(uuid.uuid4())
            
print('Concateo final apicalls')
df_apicalls = mpd.concat(apicall_df_list, ignore_index=True)
print('Salvando apicalls')
df_apicalls.to_csv(os.path.join(output_csv, f"APICALLS-{id}-parte8.csv"))
```
